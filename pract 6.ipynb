```python
# PRACTICAL 6 - AGRICULTURAL GENOMICS AND CROP IMPROVEMENT

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Step 1: Create a sample dataset
data = {
    'Marker1': [1, 0, 1, 1, 0, 1, 0, 0],
    'Marker2': [0, 1, 1, 0, 1, 1, 0, 1],
    'Marker3': [1, 1, 0, 1, 0, 0, 1, 1],
    'Yield': [3.5, 4.0, 4.5, 5.2, 3.8, 4.7, 3.2, 5.5],
    'Disease_Resistance': [8, 7, 9, 10, 6, 8, 5, 9]
}

df = pd.DataFrame(data)

# Step 2: Correlation Analysis (to find significant markers)
corr = df.corr()
print("ðŸ”¹ Correlation Matrix:\n", corr)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation of Genetic Markers with Traits")
plt.show()

# Step 3: Identify markers most correlated with yield and resistance
for marker in ['Marker1', 'Marker2', 'Marker3']:
    r_yield, _ = pearsonr(df[marker], df['Yield'])
    r_resist, _ = pearsonr(df[marker], df['Disease_Resistance'])
    print(f"Marker: {marker} | Yield corr: {r_yield:.2f} | Resistance corr: {r_resist:.2f}")

# Step 4: Conclusion
print("\nMarkers with higher positive correlation are potential candidates for crop improvement.")
```
